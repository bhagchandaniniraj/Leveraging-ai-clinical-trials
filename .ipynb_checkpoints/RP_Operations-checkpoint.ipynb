## Cell 2: Data Loading and Initial Exploration

```python
import ijson
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm
from collections import defaultdict
import gc

# Configure memory-safe settings
plt.ioff()  # Disable interactive plotting
sns.set_style('whitegrid')
CHUNK_SIZE = 1000  # Adjust based on available RAM

def process_chunk(chunk):
    """Process data chunks and collect aggregation statistics"""
    stats = {
        'sponsor': defaultdict(int),
        'phase': defaultdict(int),
        'year': defaultdict(int),
        'status': defaultdict(int)
    }
    
    for record in chunk:
        # Sponsor Class
        sponsor = record.get('protocolSection', {}).get('sponsorCollaboratorsModule', {}).get('leadSponsor', {}).get('class', 'Unknown')
        stats['sponsor'][sponsor] += 1
        
        # Phases
        phases = record.get('protocolSection', {}).get('designModule', {}).get('phases', [])
        for phase in phases:
            stats['phase'][phase] += 1
            
        # Year
        date_str = record.get('protocolSection', {}).get('statusModule', {}).get('studyFirstSubmitDate', '')
        if date_str:
            year = pd.to_datetime(date_str).year
            stats['year'][year] += 1
            
        # Status
        status = record.get('protocolSection', {}).get('statusModule', {}).get('overallStatus', 'Unknown')
        stats['status'][status] += 1
    
    return stats

def merge_stats(accumulated, new):
    """Merge chunk statistics"""
    for k in accumulated:
        for subk, v in new[k].items():
            accumulated[k][subk] += v
    return accumulated

def create_memory_safe_visualizations(stats):
    """Generate visualizations from aggregated statistics"""
    print("\n📊 Creating visualizations from aggregated data...")
    
    fig, axs = plt.subplots(2, 2, figsize=(18, 12))
    fig.suptitle('Clinical Trials Analysis (Memory Optimized)', fontsize=16)
    
    # 1. Sponsor Distribution
    sponsor_df = pd.DataFrame.from_dict(stats['sponsor'], orient='index', columns=['count']).sort_values('count', ascending=False).head(10)
    axs[0,0].pie(sponsor_df['count'], labels=sponsor_df.index, autopct='%1.1f%%')
    axs[0,0].set_title('Top 10 Sponsor Classes')
    
    # 2. Phase Distribution
    phase_df = pd.DataFrame.from_dict(stats['phase'], orient='index', columns=['count']).sort_values('count', ascending=False).head(10)
    sns.barplot(x='count', y=phase_df.index, data=phase_df, ax=axs[0,1], palette='viridis')
    axs[0,1].set_title('Top 10 Trial Phases')
    
    # 3. Yearly Trend
    year_df = pd.DataFrame.from_dict(stats['year'], orient='index', columns=['count']).sort_index()
    axs[1,0].plot(year_df.index, year_df['count'], marker='o')
    axs[1,0].set_title('Submission Timeline')
    axs[1,0].set_xlabel('Year')
    
    # 4. Status Distribution
    status_df = pd.DataFrame.from_dict(stats['status'], orient='index', columns=['count'])
    status_df.plot(kind='barh', ax=axs[1,1], legend=False)
    axs[1,1].set_title('Trial Status Distribution')
    
    plt.tight_layout()
    plt.show()

def process_large_file(filepath, split_ratio=0.7):
    """Memory-efficient processing pipeline"""
    try:
        print("🔍 Analyzing file structure...")
        with open(filepath, 'rb') as f:
            total_records = sum(1 for _ in tqdm(ijson.items(f, 'item'), desc='Counting records'))
            
        load_limit = int(total_records * split_ratio)
        processed = 0
        stats = {
            'sponsor': defaultdict(int),
            'phase': defaultdict(int),
            'year': defaultdict(int),
            'status': defaultdict(int)
        }
        
        print("\n📂 Processing data in chunks...")
        with open(filepath, 'rb') as f:
            chunk = []
            for record in tqdm(ijson.items(f, 'item'), total=total_records):
                if processed >= load_limit:
                    break
                chunk.append(record)
                processed += 1
                
                if len(chunk) >= CHUNK_SIZE:
                    chunk_stats = process_chunk(chunk)
                    stats = merge_stats(stats, chunk_stats)
                    chunk = []
                    gc.collect()
            
            if chunk:  # Process remaining records
                chunk_stats = process_chunk(chunk)
                stats = merge_stats(stats, chunk_stats)
        
        create_memory_safe_visualizations(stats)
        print("✅ Processing completed successfully!")
        
    except Exception as e:
        print(f"⚠️ Error processing file: {str(e)}")

# Execute pipeline
process_large_file("clinical_trials_cleaned.json")
```

## Cell 3: Data Cleaning and Structure Flattening

```python
# ======================
# COMPLETE CLINICAL TRIALS PIPELINE
# ======================
import ijson
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm
import gc
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Configuration
plt.ioff()
sns.set_style("whitegrid")
CHUNK_SIZE = 500  # Adjust based on RAM
FULL_DATA_PATH = "clinical_trials_full.json"
CLEANED_DATA_PATH = "clinical_trials_cleaned.json"

def create_70_30_split():
    """Create 70-30 split from full dataset with progress tracking"""
    print("🔀 Creating 70-30 data split...")
    
    # Load full data
    with open(FULL_DATA_PATH, "rb") as f:
        full_data = list(tqdm(ijson.items(f, 'item'), desc="Loading full data"))
    
    # Split data
    train_data, _ = train_test_split(full_data, test_size=0.3, random_state=42)
    
    # Save 70% cleaned data
    with open(CLEANED_DATA_PATH, 'w') as f:
        json.dump(train_data, f)
    print(f"✅ Saved {len(train_data)} records to {CLEANED_DATA_PATH}")

def stream_process_file(filepath):
    """Memory-safe JSON streaming with progress tracking"""
    try:
        print("🔢 Counting records...")
        with open(filepath, "rb") as f:
            total_records = sum(1 for _ in tqdm(ijson.items(f, "item"), 
                                              desc="Total records"))

        print("\n🔄 Processing data...")
        with open(filepath, "rb") as f:
            items = ijson.items(f, "item")
            with tqdm(total=total_records, desc="Processing") as pbar:
                while True:
                    chunk = []
                    try:
                        for _ in range(CHUNK_SIZE):
                            record = next(items)
                            chunk.append(record)
                            pbar.update(1)
                    except StopIteration:
                        pass
                    
                    if not chunk:
                        break
                        
                    yield chunk
                    del chunk
                    gc.collect()
    except Exception as e:
        print(f"🚨 Critical error: {str(e)}")
        raise

def process_chunk(chunk):
    """Process data chunk with full field retention"""
    processed = []
    for record in chunk:
        try:
            # Flatten nested structure
            flat_record = {}
            protocol = record.get("protocolSection", {})
            
            # Process modules
            modules = ["identificationModule", "statusModule", "designModule",
                      "conditionsModule", "eligibilityModule", "sponsorCollaboratorsModule"]
            for module in modules:
                mod_data = protocol.get(module, {})
                for k, v in mod_data.items():
                    flat_record[f"{module}.{k}"] = v
            
            # Age processing
            min_age = protocol.get("eligibilityModule", {}).get("minimumAge", "")
            max_age = protocol.get("eligibilityModule", {}).get("maximumAge", "")
            flat_record["minAge"] = int(min_age.split()[0]) if "Years" in min_age else np.nan
            flat_record["maxAge"] = int(max_age.split()[0]) if "Years" in max_age else np.nan
            
            # Text complexity
            criteria = protocol.get("eligibilityModule", {}).get("eligibilityCriteria", "")
            flat_record["criteriaComplexity"] = len(criteria.split())
            
            processed.append(flat_record)
        except Exception as e:
            print(f"⚠️ Record error: {str(e)}")
    return pd.DataFrame(processed)

def generate_enhanced_visuals(df):
    """Generate publication-quality visualizations"""
    print("\n📊 Creating enhanced visualizations...")
    
    # 1. Age Distribution Comparison
    plt.figure(figsize=(14, 6))
    plt.subplot(1, 2, 1)
    sns.violinplot(data=df[["minAge", "maxAge"]], palette="Set2")
    plt.title("Age Distribution Violin Plot")
    
    plt.subplot(1, 2, 2)
    sns.boxplot(data=df[["minAge", "maxAge"]], palette="Pastel1")
    plt.title("Age Distribution Boxplot")
    plt.tight_layout()
    plt.show()

    # 2. Temporal Analysis with Trend Line
    df["year"] = pd.to_datetime(df["statusModule.studyFirstSubmitDate"], errors="coerce").dt.year
    year_counts = df["year"].value_counts().sort_index()
    
    plt.figure(figsize=(12, 6))
    sns.regplot(x=year_counts.index, y=year_counts.values, order=2, 
               scatter_kws={"color": "teal"}, line_kws={"color": "red"})
    plt.title("Clinical Trial Submission Trend")
    plt.xlabel("Year")
    plt.ylabel("Number of Trials")
    plt.show()

    # 3. Sponsor-Results Relationship
    plt.figure(figsize=(12, 6))
    sponsor_results = pd.crosstab(
        df["sponsorCollaboratorsModule.leadSponsor.class"], 
        df["hasResults"]
    ).apply(lambda r: r/r.sum(), axis=1)
    sponsor_results.plot(kind="bar", stacked=True, colormap="RdYlGn")
    plt.title("Sponsor Class vs Results Availability")
    plt.xlabel("Sponsor Class")
    plt.ylabel("Proportion")
    plt.legend(title="Has Results", labels=["No", "Yes"])
    plt.show()

    # 4. Normalization Analysis
    scaler = StandardScaler()
    age_data = df[["minAge", "maxAge"]].dropna()
    if not age_data.empty:
        normalized = scaler.fit_transform(age_data)
        
        plt.figure(figsize=(12, 6))
        plt.scatter(normalized[:,0], normalized[:,1], alpha=0.6, 
                   c=df["criteriaComplexity"], cmap="viridis")
        plt.colorbar(label="Criteria Complexity")
        plt.title("Normalized Age Distribution")
        plt.xlabel("Normalized Min Age")
        plt.ylabel("Normalized Max Age")
        plt.show()

def main():
    """End-to-end workflow"""
    try:
        # Create 70-30 split if needed
        try:
            with open(CLEANED_DATA_PATH, "r") as f:
                pass
        except FileNotFoundError:
            create_70_30_split()

        # Process 70% data
        final_df = pd.DataFrame()
        for chunk in stream_process_file(CLEANED_DATA_PATH):
            chunk_df = process_chunk(chunk)
            final_df = pd.concat([final_df, chunk_df], ignore_index=True)
            del chunk_df
            gc.collect()

        # Generate analytics
        generate_enhanced_visuals(final_df)

        # Save processed data
        final_df = final_df.convert_dtypes()
        final_df.to_parquet("processed_70_data.parquet", index=False)
        
        print("\n✅ Pipeline completed successfully!")
        print(f"📦 Final dataset shape: {final_df.shape}")
        print(f"📄 Columns preserved: {len(final_df.columns)}")

    except Exception as e:
        print(f"🔥 Pipeline failed: {str(e)}")

if __name__ == "__main__":
    main()
```

## Cell 4: Data Quality Assessment and Missing Value Analysis

```python
import sys
import subprocess
import pandas as pd
import numpy as np
import json
from decimal import Decimal
from tqdm.auto import tqdm
import gc
import ijson
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns

# Configure settings
plt.ioff()
sns.set_style("whitegrid")
CHUNK_SIZE = 500  # Adjust based on RAM

class DecimalEncoder(json.JSONEncoder):
    """Custom JSON encoder for clinical trial data"""
    def default(self, obj):
        if isinstance(obj, Decimal):
            return float(obj)  # Convert Decimals to float for Parquet compatibility
        return super().default(obj)

def install_package(package):
    print(f"🔧 Installing {package}...")
    result = subprocess.run(
        [sys.executable, "-m", "pip", "install", package],
        capture_output=True,
        text=True
    )
    if result.returncode == 0:
        print(f"✅ {package} installed successfully")
    else:
        print(f"❌ Failed to install {package}: {result.stderr}")

def flatten_record(record, parent_key='', sep='_'):
    """Recursively flatten nested dictionary structure"""
    items = {}
    for k, v in record.items():
        new_key = f"{parent_key}{sep}{k}" if parent_key else k
        if isinstance(v, dict):
            items.update(flatten_record(v, new_key, sep))
        elif isinstance(v, list):
            for i, item in enumerate(v):
                if isinstance(item, dict):
                    items.update(flatten_record(item, f"{new_key}_{i}", sep))
                else:
                    items[f"{new_key}_{i}"] = item
        else:
            items[new_key] = v
    return items

def process_chunk(chunk):
    """Process a chunk of raw records"""
    processed = []
    for record in chunk:
        try:
            # Flatten nested structure
            flat_record = flatten_record(record)
            
            # Convert Decimal values and serialize complex objects
            for k, v in flat_record.items():
                if isinstance(v, Decimal):
                    flat_record[k] = float(v)
                elif isinstance(v, (dict, list)):
                    flat_record[k] = json.dumps(v, cls=DecimalEncoder)
            
            processed.append(flat_record)
        except Exception as e:
            print(f"⚠️ Error processing record: {str(e)}")
    return pd.DataFrame(processed)

def load_and_process_json(filepath):
    """Memory-safe JSON processing with progress tracking"""
    print("📥 Loading and processing JSON data...")
    try:
        file_size = Path(filepath).stat().st_size
        with open(filepath, 'rb') as f:
            items = ijson.items(f, 'item')
            with tqdm(total=file_size, desc='Processing', unit='B', unit_scale=True) as pbar:
                chunk = []
                final_df = pd.DataFrame()
                while True:
                    try:
                        record = next(items)
                        chunk.append(record)
                        pos = f.tell()
                        pbar.update(pos - pbar.n)
                        
                        if len(chunk) >= CHUNK_SIZE:
                            # Process and save chunk
                            chunk_df = process_chunk(chunk)
                            final_df = pd.concat([final_df, chunk_df], ignore_index=True)
                            chunk.clear()
                            gc.collect()
                            
                    except StopIteration:
                        if chunk:
                            chunk_df = process_chunk(chunk)
                            final_df = pd.concat([final_df, chunk_df], ignore_index=True)
                        break
                return final_df
    except Exception as e:
        print(f"❌ JSON processing failed: {str(e)}")
        return None

def generate_analytics(df):
    """Generate clinical trial insights"""
    print("\n📊 Generating analytics...")
    try:
        # 1. Study Status Distribution
        if 'protocolSection_statusModule_overallStatus' in df.columns:
            plt.figure(figsize=(10, 6))
            status_counts = df['protocolSection_statusModule_overallStatus'].value_counts()
            sns.barplot(x=status_counts.values, y=status_counts.index, palette='viridis')
            plt.title('Clinical Trial Status Distribution')
            plt.xlabel('Count')
            plt.tight_layout()
            plt.show()

        # 2. Trial Phase Analysis
        if 'protocolSection_designModule_phases' in df.columns:
            plt.figure(figsize=(10, 6))
            df['protocolSection_designModule_phases'].value_counts().plot.pie(
                autopct='%1.1f%%', startangle=90, colormap='Set3')
            plt.title('Trial Phase Distribution')
            plt.ylabel('')
            plt.show()

        # 3. Enrollment Over Time
        if 'protocolSection_statusModule_startDateStruct_date' in df.columns:
            plt.figure(figsize=(12, 6))
            dates = pd.to_datetime(df['protocolSection_statusModule_startDateStruct_date'], errors='coerce')
            dates.dropna().dt.year.value_counts().sort_index().plot(marker='o')
            plt.title('Clinical Trial Start Dates')
            plt.xlabel('Year')
            plt.ylabel('Number of Trials')
            plt.grid(True, alpha=0.3)
            plt.show()

        # 4. Sponsor Analysis
        if 'protocolSection_sponsorCollaboratorsModule_leadSponsor_class' in df.columns:
            plt.figure(figsize=(12, 6))
            sponsor_counts = df['protocolSection_sponsorCollaboratorsModule_leadSponsor_class'].value_counts().head(10)
            sns.barplot(x=sponsor_counts.values, y=sponsor_counts.index, palette='rocket')
            plt.title('Top 10 Sponsor Classes')
            plt.xlabel('Number of Trials')
            plt.tight_layout()
            plt.show()

    except Exception as e:
        print(f"❌ Analytics failed: {str(e)}")

def main():
    try:
        # Check dependencies
        try:
            import pyarrow
        except ImportError:
            install_package('pyarrow')
        
        input_file = 'clinical_trials_cleaned.json'
        output_file = 'clinical_trials_processed.parquet'

        # Process data
        df = load_and_process_json(input_file)
        if df is None or df.empty:
            print("❌ No data processed")
            return

        # Save processed data
        print("\n💾 Saving processed data...")
        df = df.convert_dtypes()
        df.to_parquet(output_file, index=False)
        print(f"✅ Data saved to {output_file}")

        # Generate analytics
        generate_analytics(df)

    except Exception as e:
        print(f"🔥 Critical failure: {str(e)}")

if __name__ == "__main__":
    main()
```

## Cell 5: Feature Engineering and Categorical Encoding

## Efficient Feature Engineering and Categorical Encoding for Large Clinical Trial Datasets
### Overview
This solution provides a robust, memory-efficient pipeline for feature engineering and categorical encoding, specifically designed for very large clinical trial datasets (5GB+). It includes progress tracking, modular transformation, and colorful visualizations for easy comparison and downstream analytics.

```python
import pandas as pd
import numpy as np
from tqdm.auto import tqdm
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure progress bars for apply
tqdm.pandas()

def extract_numerical_features(df):
    """Extract numerical features from flattened clinical trial fields with progress bars."""
    df_num = pd.DataFrame(index=df.index)

    # Age extraction
    def parse_age(age_str):
        if pd.isna(age_str) or age_str == 'Not Specified':
            return np.nan
        try:
            return float(str(age_str).split()[0])
        except:
            return np.nan

    df_num['min_age_numeric'] = df['protocolSection_eligibilityModule_minimumAge'].progress_apply(parse_age)
    df_num['max_age_numeric'] = df['protocolSection_eligibilityModule_maximumAge'].progress_apply(parse_age)

    # Text length features
    df_num['title_length'] = df['protocolSection_identificationModule_briefTitle'].astype(str).str.len()
    df_num['criteria_length'] = df['protocolSection_eligibilityModule_eligibilityCriteria'].astype(str).str.len()
    df_num['conditions_count'] = df['protocolSection_conditionsModule_conditions'].astype(str).apply(
        lambda x: x.count(',') + 1 if x != 'Unknown' else 0
    )

    # Phase encoding
    phase_map = {
        'EARLY_PHASE1': 1, 'PHASE1': 2, 'PHASE1_PHASE2': 2.5,
        'PHASE2': 3, 'PHASE2_PHASE3': 3.5, 'PHASE3': 4, 'PHASE4': 5,
        'Not Specified': 0
    }
    df_num['phase_numeric'] = df['protocolSection_designModule_phases'].map(phase_map).fillna(0).astype(float)

    return df_num

def encode_categorical_columns(df, columns):
    """Encode categorical columns using LabelEncoder with progress bar."""
    df_encoded = df.copy()
    le = LabelEncoder()
    for col in tqdm(columns, desc="Categorical Encoding"):
        if col in df_encoded.columns:
            df_encoded[f"{col}_encoded"] = le.fit_transform(df_encoded[col].astype(str))
    return df_encoded

# Load your cleaned DataFrame
df_cleaned = pd.read_parquet('clinical_trials_processed.parquet')

# Feature engineering
print("🚀 Starting feature engineering...")
df_num = extract_numerical_features(df_cleaned)

# Categorical encoding
categorical_columns = [
    'protocolSection_statusModule_overallStatus',
    'protocolSection_designModule_studyType',
    'protocolSection_eligibilityModule_sex',
    'protocolSection_sponsorCollaboratorsModule_leadSponsor_class'
]
df_encoded = encode_categorical_columns(df_cleaned, categorical_columns)

# Combine engineered features with encoded columns (no duplication)
df_engineered = pd.concat([df_encoded, df_num], axis=1)
print("✅ Feature engineering and encoding complete.")
print(f"Engineered DataFrame shape: {df_engineered.shape}")

# Save engineered data for future comparison
df_engineered.to_parquet('clinical_trials_engineered.parquet', index=False)
print("💾 Engineered features saved to clinical_trials_engineered.parquet")


def plot_engineered_features(df):
    """Create colorful, publication-ready plots for engineered features."""
    plt.figure(figsize=(18, 10))
    plt.subplot(2, 3, 1)
    sns.histplot(df['min_age_numeric'].dropna(), bins=30, color='#4ECDC4', kde=True)
    plt.title('Min Age Distribution')

    plt.subplot(2, 3, 2)
    sns.histplot(df['max_age_numeric'].dropna(), bins=30, color='#FF6B6B', kde=True)
    plt.title('Max Age Distribution')

    plt.subplot(2, 3, 3)
    sns.boxplot(x='phase_numeric', y='criteria_length', data=df, palette='Set2')
    plt.title('Criteria Length by Phase')

    plt.subplot(2, 3, 4)
    sns.violinplot(y=df['title_length'], color='#FFEAA7')
    plt.title('Title Length Distribution')

    plt.subplot(2, 3, 5)
    sns.countplot(x='conditions_count', data=df, palette='husl')
    plt.title('Number of Conditions per Trial')

    plt.subplot(2, 3, 6)
    sns.heatmap(df[['min_age_numeric', 'max_age_numeric', 'title_length', 'criteria_length', 'conditions_count', 'phase_numeric']].corr(),
                annot=True, cmap='Spectral')
    plt.title('Feature Correlation Heatmap')

    plt.tight_layout()
    plt.show()

plot_engineered_features(df_engineered)
```

Feature engineering is essential for transforming raw clinical data into machine learning-ready features . Proper encoding of categorical variables ensures algorithms can effectively process clinical trial characteristics .

## Cell 6: Data Normalization and Scaling

```python
# ===========================
# Data Normalization & Scaling
# Step 1: Data Representation and Feature Extraction
# X = [X_num, X_text] ∈ R^(d+e)
# ===========================

import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm

# Ensure progress bars for large data
tqdm.pandas()

# --- Select numerical features for normalization ---
numerical_features = [
    'min_age_numeric', 'max_age_numeric', 'title_length',
    'criteria_length', 'conditions_count', 'phase_numeric'
]

# --- Remove rows with all NaN in numerical features ---
df_for_scaling = df_engineered.dropna(subset=numerical_features, how='all').copy()

# --- Fill remaining NaN values with median (column-wise) ---
for col in numerical_features:
    df_for_scaling[col] = df_for_scaling[col].fillna(df_for_scaling[col].median())

# --- Multiple normalization techniques ---
# 1. StandardScaler (Z-score normalization)
scaler_standard = StandardScaler()
X_standard = scaler_standard.fit_transform(df_for_scaling[numerical_features])
df_standard = pd.DataFrame(X_standard, columns=[f'{col}_std' for col in numerical_features])

# 2. MinMaxScaler (0-1 normalization)
scaler_minmax = MinMaxScaler()
X_minmax = scaler_minmax.fit_transform(df_for_scaling[numerical_features])
df_minmax = pd.DataFrame(X_minmax, columns=[f'{col}_minmax' for col in numerical_features])

# 3. Robust scaling for outliers
scaler_robust = RobustScaler()
X_robust = scaler_robust.fit_transform(df_for_scaling[numerical_features])
df_robust = pd.DataFrame(X_robust, columns=[f'{col}_robust' for col in numerical_features])

print("✅ Data normalization completed using multiple techniques")
print(f"Standard scaled shape: {df_standard.shape}")
print(f"MinMax scaled shape: {df_minmax.shape}")
print(f"Robust scaled shape: {df_robust.shape}")

# --- Colorful small graphs for visual validation ---
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
sns.boxplot(data=df_for_scaling[numerical_features], palette='Set2')
plt.title('Original Features (Boxplot)')
plt.xticks(rotation=45)

plt.subplot(1, 3, 2)
sns.boxplot(data=df_standard, palette='coolwarm')
plt.title('Standard Scaled Features')
plt.xticks(rotation=45)

plt.subplot(1, 3, 3)
sns.boxplot(data=df_minmax, palette='Set3')
plt.title('MinMax Scaled Features')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

plt.figure(figsize=(18, 8))
for i, col in enumerate(numerical_features):
    plt.subplot(2, 3, i+1)
    sns.kdeplot(df_standard[f'{col}_std'], label='Standard', color='#4ECDC4', fill=True)
    sns.kdeplot(df_minmax[f'{col}_minmax'], label='MinMax', color='#FF6B6B', fill=True)
    sns.kdeplot(df_robust[f'{col}_robust'], label='Robust', color='#FFEAA7', fill=True)
    plt.title(col)
    plt.legend()
plt.tight_layout()
plt.show()
```

## Cell 7: Quartile Analysis and Statistical Assessment

```python
# ==============================
# Quartile Analysis & Visualization
# ==============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm

# --- Step 1: Quartile Analysis Function ---
def quartile_analysis(df, features):
    """Perform comprehensive quartile analysis with IQR-based outlier detection."""
    results = {}
    for feature in features:
        if feature in df.columns:
            Q1 = df[feature].quantile(0.25)
            Q2 = df[feature].quantile(0.50)  # Median
            Q3 = df[feature].quantile(0.75)
            IQR = Q3 - Q1
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR
            outliers = df[(df[feature] < lower_bound) | (df[feature] > upper_bound)]
            results[feature] = {
                'Q1': Q1,
                'Q2_Median': Q2,
                'Q3': Q3,
                'IQR': IQR,
                'Lower_Bound': lower_bound,
                'Upper_Bound': upper_bound,
                'Outlier_Count': len(outliers),
                'Outlier_Percentage': (len(outliers) / len(df)) * 100
            }
    return results

# --- Step 2: Run Quartile Analysis ---
quartile_results = quartile_analysis(df_for_scaling, numerical_features)
quartile_df = pd.DataFrame(quartile_results).T

print("=== QUARTILE ANALYSIS RESULTS ===")
print(quartile_df.round(3))

# --- Step 3: Colorful Visualizations ---
plt.figure(figsize=(18, 8))

# A. Boxplots for Each Feature
for i, feature in enumerate(numerical_features):
    plt.subplot(2, 3, i+1)
    sns.boxplot(y=df_for_scaling[feature], color=sns.color_palette('husl')[i])
    plt.title(f'Boxplot: {feature}')
    plt.ylabel('')
    plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# B. Outlier Percentage Barplot
plt.figure(figsize=(8, 4))
sns.barplot(
    x=quartile_df.index,
    y=quartile_df['Outlier_Percentage'],
    palette='mako'
)
plt.title('Outlier Percentage by Feature')
plt.ylabel('Outlier %')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# C. Quartile Summary Table Visualization
plt.figure(figsize=(10, 4))
quartile_df[['Q1', 'Q2_Median', 'Q3']].plot(kind='bar', colormap='coolwarm', ax=plt.gca())
plt.title('Quartile Summary by Feature')
plt.ylabel('Value')
plt.xticks(rotation=30)
plt.tight_layout()
plt.show()

# --- Step 4: Save Quartile Results for Future Comparison ---
quartile_df.to_csv('quartile_analysis_results.csv')
print("✅ Quartile analysis completed and results saved for future comparison.")

# --- Step 5: How to Compare in Future Cells ---
# - You can reload 'quartile_analysis_results.csv' and compare with future analyses.
# - Use the same plotting code with any new DataFrame to visually compare distributions, quartiles, and outlier rates.
```

## Cell 8: BERT Embeddings for Textual Data (Step 1 from your paper)

```python
# =========================================
# ClinicalBERT Embeddings: Full 70% Dataset, Multi-GPU/Batch Optimized, Memory-Efficient, Progress-Tracked
# =========================================

from transformers import AutoTokenizer, AutoModel
import torch
import numpy as np
import pandas as pd
from tqdm.auto import tqdm
from sklearn.decomposition import PCA
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns

# Detect and display available GPUs
gpu_count = torch.cuda.device_count()
print(f"GPUs available: {gpu_count}")
for i in range(gpu_count):
    print(f"GPU {i}: {torch.cuda.get_device_name(i)}")
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

class ClinicalBERTEmbeddings:
    def __init__(self, model_name='emilyalsentzer/Bio_ClinicalBERT', device='cuda'):
        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModel.from_pretrained(model_name)
        self.model.eval()
        self.model.to(device)
        self.device = device
    
    def get_embeddings(self, texts, batch_size=32, max_length=512):
        embeddings = []
        n = len(texts)
        for i in tqdm(range(0, n, batch_size), desc=f"Embedding texts (batch_size={batch_size})"):
            batch_texts = texts[i:i+batch_size]
            # Replace empty or nan with ""
            batch_texts = [t if isinstance(t, str) and t.strip() else "" for t in batch_texts]
            inputs = self.tokenizer(
                batch_texts,
                return_tensors='pt',
                max_length=max_length,
                truncation=True,
                padding=True
            )
            inputs = {k: v.to(self.device) for k, v in inputs.items()}
            with torch.no_grad():
                outputs = self.model(**inputs)
                batch_emb = outputs.last_hidden_state[:, 0, :].cpu().numpy()
                embeddings.append(batch_emb)
        embeddings = np.concatenate(embeddings, axis=0)
        return embeddings

# ---- Step 1: Dynamically Detect Text Columns ----
elig_col = [col for col in df_for_scaling.columns if 'eligibilityCriteria' in col][0]
title_col = [col for col in df_for_scaling.columns if 'briefTitle' in col][0]

print(f"Using eligibility column: {elig_col}")
print(f"Using title column: {title_col}")

eligibility_texts = df_for_scaling[elig_col].tolist()
title_texts = df_for_scaling[title_col].tolist()

# ---- Step 2: Multi-GPU/Batch Processing (Manual Split) ----
# If you have multiple GPUs, you can split the data and run this cell with different CUDA_VISIBLE_DEVICES settings.
# For single GPU, process all data in batches.

print("Loading Clinical BERT model...")
bert_embedder = ClinicalBERTEmbeddings(device=device)

print("Generating embeddings for eligibility criteria (full 70%)...")
eligibility_embeddings = bert_embedder.get_embeddings(eligibility_texts, batch_size=32)

print("Generating embeddings for brief titles (full 70%)...")
title_embeddings = bert_embedder.get_embeddings(title_texts, batch_size=32)

print(f"Eligibility criteria embeddings shape: {eligibility_embeddings.shape}")
print(f"Title embeddings shape: {title_embeddings.shape}")

# ---- Step 3: PCA Visualization (Small Sample) ----
sample_size = min(500, eligibility_embeddings.shape[0])
sample_idx = np.random.choice(eligibility_embeddings.shape[0], size=sample_size, replace=False)
elig_pca = PCA(n_components=2, random_state=42).fit_transform(eligibility_embeddings[sample_idx])
title_pca = PCA(n_components=2, random_state=42).fit_transform(title_embeddings[sample_idx])

plt.figure(figsize=(8, 3.5))
plt.subplot(1,2,1)
sns.scatterplot(x=elig_pca[:,0], y=elig_pca[:,1], alpha=0.5, color='#4ECDC4', s=10)
plt.title("Eligibility Embeddings (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")

plt.subplot(1,2,2)
sns.scatterplot(x=title_pca[:,0], y=title_pca[:,1], alpha=0.5, color='#FF6B6B', s=10)
plt.title("Title Embeddings (PCA)")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.tight_layout()
plt.show()

# ---- Step 4: Embedding Norm Distribution (Small Graph) ----
plt.figure(figsize=(7,3))
sns.histplot(np.linalg.norm(eligibility_embeddings[sample_idx], axis=1), bins=30, color='#45B7D1', label='Eligibility')
sns.histplot(np.linalg.norm(title_embeddings[sample_idx], axis=1), bins=30, color='#FFEAA7', label='Title', alpha=0.6)
plt.title("Embedding Norm Distribution (Sample)")
plt.xlabel("L2 Norm")
plt.legend()
plt.tight_layout()
plt.show()

# ---- Step 5: Pairwise Similarity Heatmap (Sample) ----
sim_sample_idx = np.random.choice(sample_size, size=30, replace=False)
sim_matrix = cosine_similarity(eligibility_embeddings[sample_idx][sim_sample_idx])

plt.figure(figsize=(5,4))
sns.heatmap(sim_matrix, cmap='Spectral', square=True)
plt.title("Eligibility Embedding Cosine Similarity (Sample)")
plt.tight_layout()
plt.show()

# ---- Step 6: Save Embeddings for Downstream Comparison ----
np.save('eligibility_bert_embeddings_70pct.npy', eligibility_embeddings)
np.save('title_bert_embeddings_70pct.npy', title_embeddings)
print("✅ Embeddings saved for future comparison and modeling.")

# ---- Step 7: (Optional) Multi-GPU Parallelization Instructions ----
# If you have multiple GPUs, you can split the data into N parts and run this script on each GPU
# with CUDA_VISIBLE_DEVICES set to a different GPU, then merge the resulting .npy files.
# For example, on Linux:
# CUDA_VISIBLE_DEVICES=0 python this_script.py --start 0 --end 10000
# CUDA_VISIBLE_DEVICES=1 python this_script.py --start 10000 --end 20000
# (Modify the get_embeddings method to accept start/end indices and save each chunk separately.)
```

## Cell 10: Gradient Boosting Machine Implementation (Step 3 from your paper)

```python
# Step 3: Gradient Boosting for Mixed Data
# Implementing GBM as per your paper: ŷ_GBM(X) = Σ f_k(X)

from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.ensemble import GradientBoostingClassifier

class ClinicalGBMEnsemble:
    def __init__(self):
        """Initialize multiple GBM models for ensemble"""
        # XGBoost with clinical trial optimized parameters
        self.xgb_model = XGBClassifier(
            n_estimators=100,
            max_depth=6,
            learning_rate=0.1,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42,
            eval_metric='logloss'
        )
        
        # LightGBM for efficient training
        self.lgb_model = LGBMClassifier(
            n_estimators=100,
            max_depth=6,
            learning_rate=0.1,
            subsample=0.8,
            colsample_bytree=0.8,
            random_state=42,
            verbose=-1
        )
        
        # Scikit-learn GBM
        self.sklearn_gbm = GradientBoostingClassifier(
            n_estimators=100,
            max_depth=6,
            learning_rate=0.1,
            subsample=0.8,
            random_state=42
        )
    
    def fit(self, X_train, y_train):
        """Train all GBM models"""
        print("Training XGBoost...")
        self.xgb_model.fit(X_train, y_train)
        
        print("Training LightGBM...")
        self.lgb_model.fit(X_train, y_train)
        
        print("Training Sklearn GBM...")
        self.sklearn_gbm.fit(X_train, y_train)
        
        print("✅ All GBM models trained")
    
    def predict_proba(self, X):
        """Get ensemble predictions"""
        xgb_pred = self.xgb_model.predict_proba(X)[:, 1]
        lgb_pred = self.lgb_model.predict_proba(X)[:, 1]
        sklearn_pred = self.sklearn_gbm.predict_proba(X)[:, 1]
        
        # Simple averaging ensemble
        ensemble_pred = (xgb_pred + lgb_pred + sklearn_pred) / 3
        return ensemble_pred

# Train GBM ensemble
gbm_ensemble = ClinicalGBMEnsemble()
gbm_ensemble.fit(X_train, y_train)

# Get predictions
gbm_train_pred = gbm_ensemble.predict_proba(X_train)
gbm_test_pred = gbm_ensemble.predict_proba(X_test)

print(f"✅ GBM Ensemble training completed")
print(f"Training predictions shape: {gbm_train_pred.shape}")
print(f"Test predictions shape: {gbm_test_pred.shape}")
```

## Cell 11: Ensemble Prediction (Step 4 from your paper)

```python
# Step 4: Ensemble Prediction
# ŷ_ensemble = ω * ŷ + (1-ω) * ŷ_GBM

# Train DNN for comparison
print("Training DNN model...")
train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Training loop
dnn_model.train()
for epoch in range(50):
    total_loss = 0
    for batch_X, batch_y in train_loader:
        optimizer.zero_grad()
        outputs = dnn_model(batch_X)
        loss = criterion(outputs, batch_y)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    
    if epoch % 10 == 0:
        print(f"Epoch {epoch}, Loss: {total_loss/len(train_loader):.4f}")

# Get DNN predictions
dnn_model.eval()
with torch.no_grad():
    dnn_train_pred = dnn_model(X_train_tensor).numpy().flatten()
    dnn_test_pred = dnn_model(X_test_tensor).numpy().flatten()

# Implement ensemble combination
def ensemble_prediction(dnn_pred, gbm_pred, omega=0.5):
    """
    Combine DNN and GBM predictions
    ŷ_ensemble = ω * ŷ_DNN + (1-ω) * ŷ_GBM
    """
    return omega * dnn_pred + (1 - omega) * gbm_pred

# Find optimal omega using cross-validation
from sklearn.model_selection import cross_val_score
omega_values = np.linspace(0, 1, 11)
best_omega = 0.5
best_score = 0

print("Finding optimal ensemble weight (omega)...")
for omega in omega_values:
    ensemble_pred = ensemble_prediction(dnn_train_pred, gbm_train_pred, omega)
    # Convert to binary predictions for scoring
    binary_pred = (ensemble_pred > 0.5).astype(int)
    score = accuracy_score(y_train, binary_pred)
    
    if score > best_score:
        best_score = score
        best_omega = omega
    
    print(f"Omega: {omega:.1f}, Accuracy: {score:.4f}")

print(f"\n✅ Best omega: {best_omega}, Best accuracy: {best_score:.4f}")

# Final ensemble predictions
final_train_pred = ensemble_prediction(dnn_train_pred, gbm_train_pred, best_omega)
final_test_pred = ensemble_prediction(dnn_test_pred, gbm_test_pred, best_omega)
```

The ensemble approach combines DNN and GBM predictions using an optimized weight parameter ω, as specified in your mathematical formulation . Cross-validation ensures optimal combination weights.

## Cell 13: Reinforcement Learning Framework (Step 5 from your paper)

```python
# Step 5: Reinforcement Learning for Real-Time Adaptation
# Policy updates: θ ← θ + α∇θ log π(a_t|s_t;θ)R_t

import numpy as np
from collections import namedtuple
import random

# Define clinical trial environment
Experience = namedtuple('Experience', ['state', 'action', 'reward', 'next_state'])

class ClinicalTrialEnvironment:
    """Simplified clinical trial environment for RL adaptation"""
    
    def __init__(self, n_patients=100):
        self.n_patients = n_patients
        self.current_patient = 0
        self.protocol_parameters = {
            'dosage_level': 1.0,
            'monitoring_frequency': 1.0,
            'inclusion_threshold': 0.5
        }
        self.reset()
    
    def reset(self):
        """Reset environment to initial state"""
        self.current_patient = 0
        self.protocol_parameters = {
            'dosage_level': 1.0,
            'monitoring_frequency': 1.0,
            'inclusion_threshold': 0.5
        }
        return self.get_state()
    
    def get_state(self):
        """Get current state representation"""
        return np.array([
            self.current_patient / self.n_patients,
            self.protocol_parameters['dosage_level'],
            self.protocol_parameters['monitoring_frequency'],
            self.protocol_parameters['inclusion_threshold']
        ])
    
    def step(self, action):
        """Execute action and return next state, reward"""
        # Actions: 0=decrease dosage, 1=maintain, 2=increase dosage
        if action == 0:
            self.protocol_parameters['dosage_level'] *= 0.9
        elif action == 2:
            self.protocol_parameters['dosage_level'] *= 1.1
        
        # Simulate ADE occurrence (negative reward)
        ade_probability = max(0.1, self.protocol_parameters['dosage_level'] - 0.5)
        ade_occurred = np.random.random() < ade_probability
        
        # Reward function: -1 for ADE, +0.1 for no ADE
        reward = -1.0 if ade_occurred else 0.1
        
        self.current_patient += 1
        done = self.current_patient >= self.n_patients
        
        return self.get_state(), reward, done

class PolicyGradientAgent:
    """Simple policy gradient agent for protocol adaptation"""
    
    def __init__(self, state_dim=4, action_dim=3, learning_rate=0.01):
        self.state_dim = state_dim
        self.action_dim = action_dim
        self.learning_rate = learning_rate
        
        # Simple neural network policy
        self.weights = np.random.normal(0, 0.1, (state_dim, action_dim))
        self.experiences = []
    
    def get_action_probabilities(self, state):
        """Get action probabilities using softmax"""
        logits = np.dot(state, self.weights)
        exp_logits = np.exp(logits - np.max(logits))
        return exp_logits / np.sum(exp_logits)
    
    def select_action(self, state):
        """Select action based on policy"""
        probs = self.get_action_probabilities(state)
        return np.random.choice(self.action_dim, p=probs)
    
    def store_experience(self, experience):
        """Store experience for learning"""
        self.experiences.append(experience)
    
    def update_policy(self):
        """Update policy using REINFORCE algorithm"""
        if not self.experiences:
            return
        
        states = np.array([exp.state for exp in self.experiences])
        actions = np.array([exp.action for exp in self.experiences])
        rewards = np.array([exp.reward for exp in self.experiences])
        
        # Calculate discounted returns
        returns = []
        G = 0
        for reward in reversed(rewards):
            G = reward + 0.99 * G  # gamma = 0.99
            returns.insert(0, G)
        returns = np.array(returns)
        
        # Normalize returns
        returns = (returns - np.mean(returns)) / (np.std(returns) + 1e-8)
        
        # Policy gradient update
        for i, (state, action, G) in enumerate(zip(states, actions, returns)):
            probs = self.get_action_probabilities(state)
            
            # Gradient of log policy
            grad_log_policy = np.zeros((self.state_dim, self.action_dim))
            grad_log_policy[:, action] = state
            for a in range(self.action_dim):
                if a != action:
                    grad_log_policy[:, a] = -probs[a] * state
            
            # Policy gradient update: θ ← θ + α∇θ log π(a_t|s_t;θ)R_t
            self.weights += self.learning_rate * grad_log_policy * G
        
        # Clear experiences
        self.experiences = []

# Train RL agent
print("Training Reinforcement Learning Agent for Protocol Adaptation...")

env = ClinicalTrialEnvironment()
agent = PolicyGradientAgent()

episode_rewards = []
for episode in range(100):
    state = env.reset()
    total_reward = 0
    
    while True:
        action = agent.select_action(state)
        next_state, reward, done = env.step(action)
        
        agent.store_experience(Experience(state, action, reward, next_state))
        total_reward += reward
        state = next_state
        
        if done:
            break
    
    agent.update_policy()
    episode_rewards.append(total_reward)
    
    if episode % 20 == 0:
        avg_reward = np.mean(episode_rewards[-20:])
        print(f"Episode {episode}, Average Reward: {avg_reward:.2f}")

print("✅ RL agent training completed")

# Plot learning curve
plt.figure(figsize=(10, 6))
plt.plot(episode_rewards)
plt.title('RL Agent Learning Curve')
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.grid(True)
plt.show()
```

The Reinforcement Learning implementation follows your paper's mathematical formulation for real-time protocol adaptation . The policy gradient approach enables adaptive clinical trial protocols based on observed outcomes.
